```python
target_map = {'Yes':1, 'No':0}
data["Attrition_numerical"] = data["Attrition"].apply(lambda x:target_map[x])
```

```python
data.head()
```

```python
data = data.drop(['Attrition_numerical'], axis=1)

categorical = []
for col, val in data.iteritems():
    if val.dtype == 'object':
        categorical.append(col)
        
numerical = data.columns.difference(categorical)
```

```python
print(categorical)
```

```python
print(numerical)
```

```python
data_cat = data[categorical]
data_cat = data_cat.drop(['Attrition'], axis=1)
data_cat = pd.get_dummies(data_cat)
data_num = data[numerical]
data_final = pd.concat([data_num, data_cat], axis = 1)
```

```python
target_map = {'Yes':1, 'No':0}
target = data["Attrition"].apply(lambda x:target_map[x])
```

```python
from sklearn.model_selection import train_test_split
train, test, target_train, target_val = train_test_split(data_final, target,train_size=0.80, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, log_loss, classification_report)
```

```python

seed = 0   # We set our random seed to zero for reproducibility
# Random Forest parameters
rf_params = {
    'n_jobs': -1,
    'n_estimators': 1000,
#     'warm_start': True, 
    'max_features': 0.3,
    'max_depth': 4,
    'min_samples_leaf': 2,
    'max_features' : 'sqrt',
    'random_state' : seed,
    'verbose': 0
}

rf = RandomForestClassifier(**rf_params)
rf.fit(train, target_train)
rf_predictions = rf.predict(test)
print("Accuracy score: {}".format(accuracy_score(target_val, rf_predictions)))
print("="*80)
print(classification_report(target_val, rf_predictions))
```
