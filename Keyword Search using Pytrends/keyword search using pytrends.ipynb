```python
trends = TrendReq()
```

```python
trends.build_payload(kw_list = ["Machine Learning"])
data = trends.interest_by_region()
print(data.sample(20))
```

The values are calculated on a scale from 0 to 100, where 100 is the location with the most popularity as a fraction of total searches in that location, a value of 50 indicates a location which is half as popular. A value of 0 indicates a location where there was not enough data for this term. 

```python
df = trends.trending_searches(pn='india')
df.head(20)
```

```python
df = trends.top_charts(2020, hl='en-india', tz=300, geo='GLOBAL')
df.head(10)
```

```python
keyword = trends.suggestions(keyword='machine learning')
df = pd.DataFrame(keyword)
df.drop(columns = 'mid')
```
